**EXPLORATORY DATA ANALYSIS ON STATEWISE EDUCATION SYSTEM IN INDIA**

  In this Data Analysis project, We will try to take insights from the factors affecting the education in India. We will also try to find answers to some of the interesting questions related to this.

I accessed the survey data of Statewise-Secondary Education in India of the year 2015-16. The data consists of all the related information in the context of education like literacy rate, number of schools, etc.

In order to analyse the data, I have utilized the python libraries that are available. The libraries that I used are:
*   Numpy
*   Pandas
*   Matplotlib





Therefore, I am importing all of them into my notebook.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

I read the raw data into my notebook.

***Survey_data.csv*** has the raw data that was gathered from the survey.

```python
survey_df=pd.read_csv('survey_data.csv')
```

```python
survey_df
```

```python
survey_df.statname
```

We see that the state names are not in an appropriate format. They have extra spaces in them. This should be corrected so that it doesn't create a problem while analysis.

```python
survey_df.statname=survey_df['statname'].str.strip()
survey_df.statname
```



---

Now that the data is ready, let us answer some questions related to the data.

Question 1) What are the top 10 states in terms of literacy rates?



```python
df1=survey_df[['statname','literacy_rate']]
```

```python
df1
```

As we need the list of top 10 states in terms of literacy rate, we need to sort the rows on the basis of literacy rate in descending order and take the first 10 rows.

```python
df1=df1.sort_values('literacy_rate',ascending=False).head(10)
df1
```

```python
df1.describe()
```

```python
plt.figure(figsize=(15,6))
plt.bar(df1.statname,df1.literacy_rate,color='b')
plt.xlabel("States")
plt.ylabel("Literacy Rates")
plt.title("Top 10 states in terms of literacy rate")
plt.xticks(rotation=20)
plt.yticks(np.arange(0,100,10));
```

As from the bar plot, we can say that Kerala has the highest literacy rate in India.



---

Question 2) How are the overall literacy rate, male literacy rate, female literacy rate related to each other? Is there any common trend?


```python
df2=survey_df[['statname','literacy_rate','female_literacy_rate','male_literacy_rate']].sample(20)
df2
```

```python
df2.describe()
```

```python
plt.figure(figsize=(13,6))
plt.scatter(np.arange(1,21),df2.literacy_rate,c='b')
plt.scatter(np.arange(1,21),df2.female_literacy_rate,c='#E333FF')
plt.scatter(np.arange(1,21),df2.male_literacy_rate,c='r')
plt.xticks(np.arange(1,21))
plt.yticks(np.arange(40,100,10))
plt.xlabel('States')
plt.ylabel('Literacy Rates')
plt.legend(['Overall','Female','Male']);
```

In order to compare the variables, the scatter plot is the best.

As we can observe that in all the states, the male literacy rate is the highest, and the female literacy rate is the lowest.

This gives us a valuable insight that the females are not given education as much as the males are given.



---

Question 3) Does a state having more schools have more literacy rate?

```python
df3=survey_df[['statname','schools','literacy_rate']].sort_values('literacy_rate',ascending=False)
df3['ranks']=np.arange(1,len(df3)+1)
df3.set_index('ranks')
```

We have sorted the states on basis of the literacy rates in the descending order.

That means as we go down, the literacy rates decreases.

```python
plt.figure(figsize=(15,6))
plt.plot(df3.ranks,df3.schools)
plt.xticks(np.arange(1,len(df3)+1))
plt.grid()
plt.xlabel('Ranks based on literacy rate')
plt.ylabel('No. of schools');
```

Notice that we have plotted the graph with the rank of state on X-axis and the number of schools in Y-axis.

If the notion that *'More schools equals more literacy rate'* was true, then the graph should have linear and dipping down.

But thats not really the case, therefore, we can say that the literacy rate doesn't depend on the number of schools.



---

We have seen that the girls literacy rate is comparatively lower.

Some might guess that poor facilities are the reason. Let's find out it

**Question 4) Is there a relation between female literacy rate and toilet facilities at schools?**

```python
df4=survey_df[['statname','schools','female_literacy_rate','girls_toilet_all']]
```

```python
df4
```

Let's find the percentage of schools with girls toilets

```python
df4['percentage']=(df4.girls_toilet_all/df4.schools)*100
```

```python
df4
```

```python
df4.describe()
```

```python
plt.figure(figsize=(15,6))
plt.plot(df4.index,df4.percentage)
plt.plot(df4.index,df4.female_literacy_rate)
plt.xticks(np.arange(0,len(df4)+1))
plt.grid()
plt.legend(['Percentage of schools with toilets','Female Literacy Rate']);
```

As from the graph, we can clearly see that there is absolutely no relation between the female literacy rate and the provision of girls toilets.



---
**Question 5) Does providing more girls hostels improve female literacy rate?**


Let us sort the data on the basis of female literacy rate in descending order and consider first 20 for convenience.

```python
df5=survey_df[['statname','female_literacy_rate','schools','hostel_girls_all']].sort_values('female_literacy_rate',ascending=False).head(20)
```

```python
df5
```

Taking the percentage of schools with girls hostels.

```python
df5['percentage']=(df5.hostel_girls_all/df5.schools)*100
df5.dropna(subset=['percentage'],inplace=True)
df5
```

```python
plt.figure(figsize=(15,6))
plt.plot(np.arange(1,len(df5.percentage)+1),df5.percentage)
plt.xticks(np.arange(1,len(df5.percentage)+1))
plt.xlabel('Rank of states based on female literacy rate')
plt.ylabel('Percentage of schools with girls hostel');
```

There is no linearity in the graph, and the graph should have been dipping down, if good female literacy rate is achieved by providing more girls hostels.

By which, we can conclude that there is no relation between the female literacy rate and girls hostels.
